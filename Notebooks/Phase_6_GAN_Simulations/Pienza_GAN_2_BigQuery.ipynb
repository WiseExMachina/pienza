```python
# ==============================================================================
# CELL 0: OPUS CLOUD KERNEL (Cloud-Native Bootstrap)
# ==============================================================================
# Purpose: 1. Authenticate to GCP.
#          2. Define Sovereign Project IDs.
#          3. Initialize the BigQuery Client for all subsequent operations.
#          4. Apply the Opus Lab Visual Canon.
# ==============================================================================

# --- 1. HYGIENE & IMPORTS ---
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import auth
from google.cloud import bigquery # <--- El nuevo motor

# --- 2. CLOUD CONFIGURATION (The Master Keys) ---
# Ahora usamos la jerarquía de nombres correcta:
PROJECT_ID = 'drivers-dilemma'
DATASET_REAL = 'pienza_mini'    # Contiene el Star Schema (Realidad)
DATASET_SYNTH = 'pienza_synth'  # Contiene el Manifold (Simulación)

# --- 3. AUTHENTICATION & CLIENT INITIALIZATION ---
print("🔐 Iniciando Solicitud de Soberanía Cloud (Autenticación GCP)...")
try:
    auth.authenticate_user()
    print("✅ Autenticación de Usuario Exitosa.")

    client = bigquery.Client(project=PROJECT_ID)
    print(f"🚀 Cliente BigQuery Activo en proyecto: {PROJECT_ID}")

except Exception as e:
    print(f"❌ FALLO CRÍTICO DE CONEXIÓN: {e}")
    raise

# --- 4. VISUAL CANON (OPUS LAB THEME) ---
OPUS_PURPLE = '#440154'
OPUS_TEAL   = '#21918c'
OPUS_GREY   = '#FAFAFA'
OPUS_TEXT   = '#121212'

sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'figure.facecolor': OPUS_GREY,
    'axes.facecolor': OPUS_GREY,
    'text.color': OPUS_TEXT,
    'xtick.color': '#333333',
    'ytick.color': '#333333',
    'axes.edgecolor': '#DDDDDD',
    'grid.color': '#E0E0E0',
    'font.family': 'sans-serif',
    'axes.titlecolor': OPUS_PURPLE,
    'axes.titleweight': 'bold',
    'figure.titlesize': 24,
    'figure.titleweight': 'bold'
})

print("✅ Visual Identity Loaded: Opus Lab (Cloud Mode).")
print(f"🔗 Ready to operate on {DATASET_REAL} (Reality) and {DATASET_SYNTH} (Simulation).")
print("\n--- SYSTEM READY ---")
```

```python
# ==============================================================================
# CELL 1 (REVISADA V2): COPIA INTELIGENTE (EXTERNAL -> NATIVE)
# ==============================================================================

SOURCE_DATASET = 'opus_phase_4'
SOURCE_TABLE = 'gan_training_set'
DEST_DATASET = DATASET_SYNTH  # 'pienza_synth'
DEST_TABLE = 'gan_training_set_v1_reference'

# --- A. GARANTIZAR INFRAESTRUCTURA (Re-ejecución) ---
def prepare_destination_dataset():
    print(f"🏗️ Verificando y creando Dataset: {DEST_DATASET}")
    dataset_ref = bigquery.DatasetReference(PROJECT_ID, DEST_DATASET)
    try:
        client.get_dataset(dataset_ref)
    except Exception:
        ds = bigquery.Dataset(dataset_ref)
        ds.location = "US"
        client.create_dataset(ds)

# --- B. EJECUTAR LA CONVERSIÓN Y COPIA (CTAS) ---
def copy_training_base_fixed_v2():
    print(f"\n🔄 Iniciando COPIA/CONVERSIÓN de External Table a Tabla Nativa...")

    source_id = f"`{PROJECT_ID}.{SOURCE_DATASET}.{SOURCE_TABLE}`"
    dest_id = f"{PROJECT_ID}.{DEST_DATASET}.{DEST_TABLE}"

    query = f"""
    CREATE OR REPLACE TABLE `{dest_id}`
    AS
    SELECT * FROM {source_id}
    """

    try:
        # Ejecutar el query de creación de tabla (CTAS)
        job = client.query(query)
        job.result()

        print(f"✅ CONVERSIÓN COMPLETADA. External Link ({SOURCE_TABLE}) convertido a Tabla Nativa ({DEST_TABLE}).")

    except Exception as e:
        print(f"❌ FALLO CRÍTICO EN CONVERSIÓN CTAS: {e}")

# --- EJECUCIÓN SECUENCIAL ---
prepare_destination_dataset() # Garantizamos el contenedor
copy_training_base_fixed_v2() # Ejecutamos la copia nativa
```

```python
# ==============================================================================
# CELL 2: DESPLIEGUE DEL MANIFOLD V2
# ==============================================================================
# Purpose: Despliega el Parquet final del GAN (V2) al dataset pienza_synth.
# ==============================================================================

# PASTE EL ID DE TU NUEVO ARCHIVO PARQUET AQUÍ
NEW_MANIFOLD_FILE_ID = '1wEffNLaLQQcIrV7Vx4PIEI5nxyE2GPD4'

TABLE_SYNTH = 'synthetic_manifold'

def deploy_synthetic_universe_final():
    print(f"\n🌉 Iniciando conexión del Manifold V2 a {DATASET_SYNTH}...")

    if 'TU_NUEVO_FILE_ID_AQUÍ' in NEW_MANIFOLD_FILE_ID:
        print("❌ ERROR: Por favor, actualiza NEW_MANIFOLD_FILE_ID con el ID real del archivo Parquet.")
        return

    table_ref = client.dataset(DATASET_SYNTH).table(TABLE_SYNTH)
    uri = f'https://drive.google.com/open?id={NEW_MANIFOLD_FILE_ID}'

    # Configuración de conexión (Tabla Externa)
    ext_config = bigquery.ExternalConfig("PARQUET")
    ext_config.source_uris = [uri]
    ext_config.autodetect = True

    table = bigquery.Table(table_ref)
    table.external_data_configuration = ext_config

    try:
        # Reemplazamos la tabla si ya existe para asegurar que el puntero es correcto
        client.delete_table(table_ref, not_found_ok=True)
        client.create_table(table)
        print(f"✅ MANIFOLD V2 DESPLEGADO Y CONECTADO.")

    except Exception as e:
        print(f"❌ ERROR EN EL DESPLIEGUE: {e}")

# EJECUTAR
deploy_synthetic_universe_final() # Descomenta cuando pegues el ID
```
