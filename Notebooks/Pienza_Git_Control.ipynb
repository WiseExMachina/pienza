# Celda para Bajar TEX a /Scientific

```python
# ==============================================================================
# OPERACIÓN: TIERRA QUEMADA (BORRAR TODO -> DESCARGAR FRESCO)
# ==============================================================================
import os
import shutil
import urllib.parse
import subprocess
from google.colab import userdata

# 1. SEGURIDAD DE ENTORNO
# Nos movemos a la raíz para no estar parados sobre la carpeta que vamos a explotar
os.chdir("/content")

# 2. CONFIGURACIÓN
RUTA_DESTINO = "/content/drive/MyDrive/_Pienza/Github_Repository/pienza/The_Pienza_Papers/Pienza_Papers_Scientific"

print("🔥 INICIANDO PROTOCOLO DE TIERRA QUEMADA...")

# 3. AUTENTICACIÓN
try:
    raw_token = userdata.get('ov_pass')
    raw_url = userdata.get('ov_url_base')
    clean_base = raw_url.replace("https://", "").replace("http://", "").strip("/")
    safe_token = urllib.parse.quote(raw_token, safe='')
    GIT_URL_FINAL = f"https://git:{safe_token}@{clean_base}"
except Exception as e:
    print(f"🔴 ERROR CREDENCIALES: {e}")
    raise

# 4. ELIMINACIÓN TOTAL (SCORCHED EARTH)
if os.path.exists(RUTA_DESTINO):
    print(f"🗑️ Eliminando carpeta existente: {os.path.basename(RUTA_DESTINO)}...")
    shutil.rmtree(RUTA_DESTINO) # Adiós. Se borró todo.
else:
    print("⚠️ La carpeta no existía. Creando desde cero...")

# 5. DESCARGA DIRECTA (CLONADO)
print("📡 Descargando copia fresca desde Overleaf...")

try:
    # Al no existir la carpeta (porque la borramos), Git clone la crea y descarga ahí directo.
    subprocess.run(["git", "clone", GIT_URL_FINAL, RUTA_DESTINO], check=True, capture_output=True)
    print("✅ Descarga completada.")
except subprocess.CalledProcessError as e:
    print(f"💥 ERROR CRÍTICO: Falló la descarga.\n{e.stderr.decode('utf-8')}")
    print("☠️ ALERTA: Tu carpeta local ha sido eliminada y la descarga falló. Restaura desde Overleaf si es necesario.")
    raise Exception("Misión Fallida.")

# 6. LIMPIEZA TÁCTICA (ELIMINAR RASTROS DE GIT)
# Esto es vital. Borramos la carpeta .git que vino de Overleaf para que no choque con tu Repo Pienza.
git_folder_overleaf = os.path.join(RUTA_DESTINO, ".git")
if os.path.exists(git_folder_overleaf):
    shutil.rmtree(git_folder_overleaf)
    print("🧹 Carpeta .git de Overleaf eliminada (Integración limpia asegurada).")

print("\n🏆 OPERACIÓN FINALIZADA.")
print("   Tienes una copia 100% fresca de Overleaf en tu Drive.")
print("👉 EJECUTA TU 'GIT COMMIT' UNIVERSAL AHORA.")
```

# CELDA PARA MATAR TOXICIDAD DE WIDGETS (ANTES DE CADA COMMIT):

```python
import json
import glob
import os

# ==============================================================================
# CONFIGURACIÓN DE RUTA: TU REPOSITORIO
# ==============================================================================
REPO_ROOT = "/content/drive/MyDrive/_Pienza/Github_Repository/pienza"

def sanitize_notebooks(directory_path):
    print(f"🕵️‍♂️ Escaneando sector: {directory_path} en busca de widgets tóxicos...")

    # Buscar todos los archivos .ipynb (recursivamente en carpetas)
    files = glob.glob(f"{directory_path}/**/*.ipynb", recursive=True)

    cleaned_count = 0

    for file_path in files:
        try:
            # 1. Cargar el JSON del notebook
            with open(file_path, 'r', encoding='utf-8') as f:
                nb = json.load(f)

            # 2. DETECTAR INFECCIÓN: ¿Tiene metadatos de widgets?
            is_infected = 'widgets' in nb.get('metadata', {})

            # 3. CIRUGÍA: Extirpar la sección 'widgets'
            if is_infected:
                del nb['metadata']['widgets']

                # Guardar el paciente curado
                with open(file_path, 'w', encoding='utf-8') as f:
                    json.dump(nb, f, indent=1)

                print(f"   💊 DESINFECTADO: {os.path.basename(file_path)}")
                cleaned_count += 1

        except Exception as e:
            print(f"   ⚠️ Error leyendo {os.path.basename(file_path)}: {e}")

    if cleaned_count == 0:
        print("✅ TODO LIMPIO. Ningún archivo requirió cirugía.")
    else:
        print(f"✅ OPERACIÓN COMPLETADA: {cleaned_count} archivos limpiados.")
        print("   -> Ahora es seguro hacer Commit y Push. GitHub mostrará el historial.")

# EJECUCIÓN INMEDIATA
sanitize_notebooks(REPO_ROOT)
```
