```python
import os # Asegúrate de que os está importado si no lo está ya

# --- CONFIGURACIÓN DE LA RUTA DEL ARCHIVO CSV ---
csv_file_path = '/content/drive/MyDrive/casos_confirmados.csv' # Asegúrate de que esta ruta sea correcta

# --- CONFIGURACIÓN DE CODIFICACIÓN ---
file_encoding = 'utf-8'

# --- VARIABLES PARA LOS CÁLCULOS ---
total_age = 0
count_patients = 0
gender_counts = {'MALE': 0, 'FEMALE': 0, 'NO INFO': 0}
age_column_index = -1
gender_column_index = -1
line_count = 0 # << CORRECCIÓN 1: Inicializar line_count aquí fuera

print(f"Intentando abrir y procesar el archivo: '{csv_file_path}'")

if not os.path.exists(csv_file_path):
    print(f"Error: El archivo '{csv_file_path}' no se encontró.")
else:
    try:
        with open(csv_file_path, mode='r', encoding=file_encoding) as infile:
            # --- LECTURA DE ENCABEZADOS A MANO ---
            header_line = infile.readline().strip()
            headers = header_line.split(',')
            print(f"Encabezados detectados: {headers}")

            # Identificar los índices de las columnas correctas
            try:
                # << CORRECCIÓN 2: Usar los nombres de columna correctos 'Age' y 'Sex'
                age_column_index = [i for i, h in enumerate(headers) if h.strip().upper() == 'AGE'][0]
                gender_column_index = [i for i, h in enumerate(headers) if h.strip().upper() == 'SEX'][0]
                print(f"Índice de columna 'Age': {age_column_index}")
                print(f"Índice de columna 'Sex': {gender_column_index}")

            except IndexError:
                print("\nError: No se pudieron encontrar las columnas 'Age' o 'Sex' en los encabezados.")
                headers = None

            # --- ITERACIÓN Y PROCESAMIENTO MANUAL ---
            if headers is not None:
                for line in infile:
                    line_count += 1
                    if line_count % 10000 == 0:
                        print(f"Procesadas {line_count} líneas...")

                    row = line.strip().split(',')

                    try:
                        if len(row) > max(age_column_index, gender_column_index):
                            # Recuperar la edad
                            age_str = row[age_column_index]
                            try:
                                age = int(age_str)
                                if age >= 0:
                                    total_age += age
                                    count_patients += 1
                            except ValueError:
                                pass # Ignorar si la edad no es un número

                            # << CORRECCIÓN 3: Ajustar la lógica de género a los datos en inglés 'Male'/'Female'
                            gender_raw = row[gender_column_index].strip().upper()
                            if gender_raw == 'MALE':
                                gender_counts['MALE'] += 1
                            elif gender_raw == 'FEMALE':
                                gender_counts['FEMALE'] += 1
                            else:
                                gender_counts['NO INFO'] += 1

                    except Exception as e:
                        print(f"Error procesando línea {line_count + 1}: {e}.")

        print(f"\nProcesamiento completado. Total de líneas leídas (excluyendo encabezados): {line_count}")

    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
```

```python
# --- CÁLCULO DE EDAD PROMEDIO FINAL ---

average_age = 0 # Valor por defecto

if count_patients > 0:
    average_age = total_age / count_patients
    print(f"Edad promedio calculada a partir de {count_patients} pacientes válidos.")
    print(f"¿Cuál es la edad promedio de los pacientes? \n   La edad promedio es: {average_age:.2f} años")
else:
    print("No se encontraron pacientes con edad válida para calcular el promedio.")
    print("¿Cuál es la edad promedio de los pacientes? \n   No se pudo calcular la edad promedio.")
```

```python
import os

# --- CONFIGURACIÓN DE LA RUTA DEL ARCHIVO CSV ---
csv_file_path = '/content/drive/MyDrive/casos_confirmados.csv'

# --- CONFIGURACIÓN DE CODIFICACIÓN ---
file_encoding = 'utf-8'

# --- VARIABLES PARA LOS CÁLCULOS ---
total_age = 0
count_patients_age = 0
gender_counts = {'MASCULINO': 0, 'FEMENINO': 0, 'SIN ESPECIFICAR': 0} # <-- Usamos los nombres en español
age_column_index = -1
gender_column_index = -1
line_count = 0

print(f"Intentando abrir y procesar el archivo: '{csv_file_path}'")

if not os.path.exists(csv_file_path):
    print(f"Error: El archivo '{csv_file_path}' no se encontró.")
else:
    try:
        with open(csv_file_path, mode='r', encoding=file_encoding) as infile:
            header_line = infile.readline().strip()
            headers = header_line.split(',')
            print(f"Encabezados detectados: {headers}")

            try:
                age_column_index = [i for i, h in enumerate(headers) if h.strip().upper() == 'AGE'][0]
                gender_column_index = [i for i, h in enumerate(headers) if h.strip().upper() == 'SEX'][0]
                print(f"Índice de columna 'Age': {age_column_index}")
                print(f"Índice de columna 'Sex': {gender_column_index}")
            except IndexError:
                print("\nError: No se pudieron encontrar las columnas 'Age' o 'Sex'.")
                headers = None

            if headers is not None:
                for line in infile:
                    line_count += 1
                    row = line.strip().split(',')

                    try:
                        if len(row) > max(age_column_index, gender_column_index):
                            # --- Recuperar y procesar la EDAD ---
                            age_str = row[age_column_index]
                            try:
                                age = int(age_str)
                                if 0 <= age <= 120:
                                    total_age += age
                                    count_patients_age += 1
                            except ValueError:
                                pass

                            # --- Recuperar y procesar el GÉNERO (LÓGICA FINALMENTE CORRECTA) ---
                            gender_raw = row[gender_column_index].strip().upper()
                            if gender_raw == 'FEMENINO':
                                gender_counts['FEMENINO'] += 1
                            elif gender_raw == 'MASCULINO':
                                gender_counts['MASCULINO'] += 1
                            else:
                                gender_counts['SIN ESPECIFICAR'] += 1

                    except Exception as e:
                        pass

        print(f"\nProcesamiento completado. Total de líneas leídas: {line_count}")

    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")

# Determinación del género predominante
print(f"Respuesta: ¿Hay más hombres o mujeres contagiados?")
print(f"Conteo de Géneros: {gender_counts}")

if gender_counts['MASCULINO'] > gender_counts['FEMENINO']:
    print("Hay más hombres contagiados.")
elif gender_counts['FEMENINO'] > gender_counts['MASCULINO']:
    print("Hay más mujeres contagiadas.")
else:
    if gender_counts['MASCULINO'] == 0 and gender_counts['FEMENINO'] == 0:
        print("No se encontraron casos con género especificado.")
    else:
        print("Hay una cantidad igual de hombres y mujeres contagiados.")
```
