```python
# ==============================================================================
# OPUS CLOUD FOUNDRY: IGNITION
# ==============================================================================
# Status: Founding Commit - January 14, 2026
# Environment: BigQuery Studio / us-central1
# Purpose: Establish Sovereign Connectivity to the Star Schema.
# ==============================================================================

from google.cloud import bigquery
import pandas as pd

# --- 1. SOVEREIGN CONFIGURATION ---
PROJECT_ID = 'drivers-dilemma'
DATASET_ID = 'opus_mini'

# --- 2. INITIALIZE NATIVE CLIENT ---
# No manual auth needed; BQ Studio uses the underlying Service Account.
client = bigquery.Client(project=PROJECT_ID)

print(f"🚀 Opus Cloud Foundry Online.")
print(f"📡 Project: {PROJECT_ID} | Dataset: {DATASET_ID}")

# --- 3. THE "VOBO" PROBE ---
# Verifying that the Cloud Star Schema is readable.
test_query = f"SELECT COUNT(*) as total_offers FROM `{PROJECT_ID}.{DATASET_ID}.offers`"
try:
    df_check = client.query(test_query).to_dataframe()
    print(f"✅ Connectivity Verified: {df_check.iloc[0,0]} offers are visible.")
except Exception as e:
    print(f"❌ Connectivity Failed: {e}")
```
