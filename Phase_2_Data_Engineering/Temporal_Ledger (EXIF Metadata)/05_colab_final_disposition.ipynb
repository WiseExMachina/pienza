```python
# ==============================================================================
# SCRIPT:         colab_final_disposition.ipynb
# OPERATION:      Final Disposition (Executive Order)
# PURPOSE:        Classifies ALL "Requires Investigation" assets with a single,
#                 default status and produces the final auditable report.
#                 This script acts as the implementation of a strategic
#                 decision to halt manual review.
# ==============================================================================
import pandas as pd
import os
from google.colab import auth
from google.colab import drive

print("--- Executing Final Disposition Protocol ---")

# --- Step 1: Authentication & Mounting ---
print("Authenticating user for Google services...")
auth.authenticate_user()
print("✅ Authentication successful.")
print("\nMounting Google Drive...")
drive.mount('/content/drive')
print("✅ Google Drive mounted successfully.")

# --- Step 2: Configuration ---
# Input: The report generated by the initial analysis notebook.
INPUT_REPORT_PATH = '/content/drive/My Drive/Opus_1_101001/Database/03_analysis/asset_disposition_report.csv'

# Output: The final, definitive report.
OUTPUT_FOLDER_PATH = '/content/drive/My Drive/Opus_1_101001/Database/03_analysis/'
FINAL_OUTPUT_FILENAME = 'disposition_report_FINAL.csv'

# The default status that will be applied to all un-logged, non-duplicate images.
DEFAULT_STATUS = "Dropped (Assumed Non-Critical per Executive Order)"

try:
    # --- Step 3: Load Data ---
    print(f"\nLoading source report from:\n > '{INPUT_REPORT_PATH}'")
    if not os.path.exists(INPUT_REPORT_PATH):
        raise FileNotFoundError("The initial disposition report was not found. Please ensure the 'colab_asset_disposition' notebook has been run successfully.")

    df_final = pd.read_csv(INPUT_REPORT_PATH)

    # --- Step 4: Apply Bulk Classification ---
    print(f"\nApplying default status ('{DEFAULT_STATUS}') to all assets marked 'Requires Investigation'...")

    # Create a new column 'Final_Disposition' for the final classification.
    # First, copy the initial classification to the new column.
    df_final['Final_Disposition'] = df_final['Classification']

    # Now, find all rows that require investigation and update their status in the new column.
    df_final.loc[
        df_final['Classification'] == 'Requires Investigation',
        'Final_Disposition'
    ] = DEFAULT_STATUS

    # --- Step 5: Final Report Generation ---
    print("\n--- Final Disposition Summary ---")
    print(df_final['Final_Disposition'].value_counts().to_string())

    final_report_path = os.path.join(OUTPUT_FOLDER_PATH, FINAL_OUTPUT_FILENAME)
    df_final.to_csv(final_report_path, index=False)

    print(f"\n✅ Operation Complete. The final, fully-classified disposition report has been saved to your Google Drive at:")
    print(f"   > '{final_report_path}'")

except FileNotFoundError as e:
    print(f"\n❌ ERROR: {e}")
except Exception as e:
    print(f"\n❌ An unrecoverable error occurred: {e}")
```
